```python
Top_Artists = df.groupby('Artist')['Total Hours Streamed (Millions)'].sum().sort_values(ascending= False).head(10)
Top_Artists
```

```python
Top_Albums = df.groupby("Album")["Total Hours Streamed (Millions)"].sum().sort_values(ascending = False).head(10)
Top_Albums
```

```python
import matplotlib.pyplot as plt

Top_Artists.plot(kind='barh', title='Top 10 Artists by Hours Streamed (2024)')
plt.ylabel('Total Hours Streamed')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

Top_Albums.plot(kind='barh', title='Top 10 Albums by Hours Streamed (2024)')
plt.ylabel('Total Hours Streamed')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
Top_Artists_by_Country = df.groupby(['Country', 'Artist'])['Total Hours Streamed (Millions)'].sum().reset_index()
Top_Artists_by_Country
```

```python
Top_per_Country = Top_Artists_by_Country.sort_values(['Country', 'Total Hours Streamed (Millions)'], ascending=[True, False])
Top_per_Country
```

```python
df.plot.scatter(x='Monthly Listeners (Millions)', y='Total Streams (Millions)',
                title='Monthly Listeners vs Total Streams')
```

Future Potential of the Spotify 2024 Global Streaming Dataset
The Spotify 2024 Global Streaming Dataset offers rich insights into global music consumption trends. With detailed information on countries, artists, genres, streaming hours, listener counts, and platform types, this dataset serves as a valuable resource for:

Predictive Modeling: Future listening trends could be forecasted based on historical streaming data.

Market Expansion: Record labels and artists can identify emerging markets and fan bases by analyzing growth patterns across regions.

Genre Evolution Analysis: Comparing current and past datasets could reveal how global genre preferences shift over time.

Cultural Insight: This data could help researchers understand the cultural impact of global music trends and their variation by location.

Marketing Strategy: Streaming platforms and promoters can tailor campaigns and product releases by pinpointing where certain artists or genres are gaining traction.

As music continues to globalize, this dataset could play a crucial role in shaping strategies for artists, marketers, platforms, and researchers alike — turning raw data into actionable, trend-driven decisions.
